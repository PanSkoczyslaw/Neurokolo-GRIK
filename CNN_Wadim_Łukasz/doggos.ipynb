```python
ds_path = '/content/drive/MyDrive/images pjeski'
def facialazing(img_folder,IMG_HEIGHT, IMG_WIDTH):
  # dirs = os.listdir(path)
  img_data_array=[]
  class_name=[]
  for dir1 in os.listdir(img_folder):
        for file in os.listdir(os.path.join(img_folder, dir1)):
          # filename, ext = os.path.splitext(os.path.basename(path))
          image_path= os.path.join(img_folder, dir1,  file)
          img = cv2.imread( image_path, cv2.COLOR_BGR2RGB)

          # img = cv2.cvtColor(img[i], cv2.COLOR_BGR2RGB)
          img_result = img.copy()
          dets = detector(img, upsample_num_times=1)
          dets_result = detector(img_result, upsample_num_times = 1)
          for i, d in enumerate(dets):
            x1, y1 = d.rect.left(), d.rect.top()
            x2, y2 = d.rect.right(), d.rect.bottom()
          
            crop_img = img[y1:y2, x1:x2]
            try:
              crop_img=cv2.resize(crop_img, (IMG_HEIGHT, IMG_WIDTH),interpolation = cv2.INTER_AREA)
              crop_img=np.array(crop_img)
              crop_img = crop_img.astype('float32')
              crop_img /= 255
              img_data_array.append(crop_img)
              class_name.append(dir1)
              # doggos_preprocessed = tf.keras.applications.vgg16.preprocess_input(crop_img)
              # img_data_array.append(doggos_preprocessed)
              # class_name.append(dir1)
            except Exception as e:
              pass
           
          filtered = img_result[0:IMG_HEIGHT, 0:IMG_WIDTH]
          try:
            filtered=cv2.resize(filtered, (IMG_HEIGHT, IMG_WIDTH),interpolation = cv2.INTER_AREA)
            filtered=np.array(filtered)
            filtered = filtered.astype('float32')
            filtered /= 255
            img_data_array.append(filtered)
            class_name.append(dir1)
              # doggos_preprocessed = tf.keras.applications.vgg16.preprocess_input(crop_img)
              # img_data_array.append(doggos_preprocessed)
              # class_name.append(dir1)
          except Exception as e:
            pass 
            # cv2.waitKey(0)
            # img_array.append(crop_img)
  return img_data_array, class_name
```

```python
doggos_train_X = np.asarray(doggos_train_X)
doggos_test_X = np.asarray(doggos_test_X)
doggos_train_Y = np.asarray(list(map(lambda l : target_dict[l], doggos_train_Y)))
doggos_test_Y = np.asarray(list(map(lambda l : target_dict[l], doggos_test_Y)))
```

```python
from keras.applications.vgg16 import VGG16

from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras import models, layers
```

```python
baseModel = VGG16(include_top=False, weights="imagenet", input_shape=doggos_train_X[0].shape)
baseModel.trainable = False
```

```python
baseModel.summary()
```

```python
flatten_layer = layers.Flatten()
prediction_layer = layers.Dense(4,activation='softmax')
dense_layer_1 = layers.Dense(80, activation='relu')
dense_layer_2 = layers.Dense(40, activation='relu')

model = models.Sequential([
    baseModel,
    flatten_layer,
    dense_layer_1,
    dense_layer_2,
    prediction_layer
])
```

```python
model.compile(optimizer='adam', loss="sparse_categorical_crossentropy", metrics=['accuracy'])
es = EarlyStopping(monitor='val_accuracy', mode='max', patience=20, restore_best_weights=True)
```

```python
model.fit(doggos_train_X, doggos_train_Y, epochs = 100, validation_split=0.2, batch_size=32, callbacks=[es])
```
